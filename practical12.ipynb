### Multiple clusters from column data entries  

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
```

```python
# 1. Load CSV file
# df = pd.read_csv('data.csv')

# using iris instead of loading csv file
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
# Step 2: Select column to cluster (e.g., 'petal length (cm)')
column_to_cluster = 'petal length (cm)'
data = df[[column_to_cluster]]
```

```python
# Step 3: Choose number of clusters (k)
k = 3
```

```python
# Step 4: Apply K-Means
kmeans = KMeans(n_clusters=k, random_state=42)
df['Cluster'] = kmeans.fit_predict(data)
```

```python
# Step 5: Output the DataFrame with cluster labels
print(df[[column_to_cluster, 'Cluster']].head(10))
```

```python
# Step 6: Plot the clusters
plt.scatter(df.index, df[column_to_cluster], c=df['Cluster'], cmap='viridis')
plt.xlabel("Index")
plt.ylabel(column_to_cluster)
plt.title("K-Means Clustering on Petal Length (Iris Dataset)")
plt.colorbar(label='Cluster')
plt.show()
```
