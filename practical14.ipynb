###  To use healthcare dataset and form scatter plot with observed statistical measures. 

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
# Step 1: Load healthcare dataset
data = load_breast_cancer()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['label'] = data.target
```

```python
# Step 2: Select two features for scatter plot
X = df[['mean radius', 'mean texture']]
y = df['label']
```

```python
# Step 3: Split into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Step 4: Choose k
k = 5
```

```python
# Step 5: Train k-NN model
knn = KNeighborsClassifier(n_neighbors=k)
knn.fit(X_train, y_train)
```

```python
# Step 6: Predict
y_pred = knn.predict(X_test)
```

```python
# Step 7: Accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")
```

```python
# Step 8: Scatter Plot with true labels
plt.figure(figsize=(8, 6))
plt.scatter(X_test['mean radius'], X_test['mean texture'], c=y_pred, cmap='coolwarm', edgecolors='k')
plt.xlabel('Mean Radius')
plt.ylabel('Mean Texture')
plt.title(f'k-NN Classification (k={k}) - Predicted Clusters\nAccuracy: {accuracy:.2f}')
plt.colorbar(label='Predicted Label')
plt.grid(True)
plt.show()
```
