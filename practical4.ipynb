### Data Generation and Preprocessing for Regression Models

```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
# Generate dataset 
X = np.linspace(-5, 5, 100) 
y = np.sin(X) + np.random.uniform(-0.2, 0.2, X.shape)
```

```python
# Transforming data into polynomial features 
def generate_poly_features(X, degree): 
    return np.array([X**i for i in range(degree + 1)]).T 
X_poly = generate_poly_features(X, degree=3)
```

```python
# Visualization 
plt.scatter(X, y, color='blue', label="Generated data") 
plt.xlabel("X") 
plt.ylabel("y") 
plt.title("Generated Data for Regression") 
plt.legend() 
plt.show()
```
