###  Implementation of Logistic Regression and Evaluation of Cost Function and Gradient Descent

```python
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn.datasets import make_classification 
```

```python
# Generate synthetic dataset 
# Reduced n_redundant to 0 
X, y = make_classification(n_samples=100, n_features=2, n_classes=2, random_state=1, 
n_redundant=0) 
y = y.reshape(-1, 1) 
m, n = X.shape
```

```python
# Initialize parameters 
theta = np.zeros((n, 1)) 
learning_rate = 0.1 
iterations = 1000 
```

```python
# Sigmoid function 
def sigmoid(z): 
    return 1 / (1 + np.exp(-z))
```

```python
# Cost function 
def compute_cost(X, y, theta): 
    h = sigmoid(X @ theta) 
    cost = -(1/m) * np.sum(y * np.log(h) + (1 - y) * np.log(1 - h)) 
    return cost
```

```python
# Gradient descent 
cost_history = [] 
for _ in range(iterations): 
    h = sigmoid(X @ theta) 
    gradient = (1/m) * (X.T @ (h - y)) 
    theta -= learning_rate * gradient 
    cost_history.append(compute_cost(X, y, theta))
```

```python
# Plot cost 
plt.plot(cost_history) 
plt.xlabel('Iterations') 
plt.ylabel('Cost') 
plt.title('Cost function reduction over iterations') 
plt.show() 
print("Final Cost:", cost_history[-1])
```
