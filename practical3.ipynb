### Polynomial Regression and Comparison with Linear Regression 

```python
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn.preprocessing import PolynomialFeatures 
from sklearn.linear_model import LinearRegression
```

```python
# Generate dataset 
X = np.linspace(0, 10, 50).reshape(-1, 1) 
y = 3 * X**2 - 2 * X + 4 + np.random.randn(50, 1) * 5 
```

```python
# Polynomial regression 
poly = PolynomialFeatures(degree=2) 
X_poly = poly.fit_transform(X) 
model = LinearRegression() 
model.fit(X_poly, y) 
y_poly_pred = model.predict(X_poly) 
```

```python
# Linear regression 
linear_model = LinearRegression() 
linear_model.fit(X, y) 
y_linear_pred = linear_model.predict(X)
```

```python
# Visualization 
plt.scatter(X, y, color='blue', label="Noisy data") 
plt.plot(X, y_poly_pred, color='red', label="Polynomial Regression") 
plt.plot(X, y_linear_pred, color='green', label="Linear Regression") 
plt.legend() 
plt.xlabel("X") 
plt.ylabel("y") 
plt.title("Polynomial vs Linear Regression") 
plt.show()
```
